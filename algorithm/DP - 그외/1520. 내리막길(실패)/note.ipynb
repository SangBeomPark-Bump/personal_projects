```python
cnt = 0
sad_cnt = 0
visited = [[0] * N for _ in range(M)]

def dfs(y, x):
    global cnt, sad_cnt
    if x == N-1 and y == M-1:
        print('앗! 종료에요! 카운터야 올라랏!!')
        cnt +=1
        return

    dxs = [-1, 1, 0, 0]
    dys = [0, 0, -1, 1]
    for dx, dy in zip(dxs, dys):
        nx = x +dx
        ny = y +dy
        if 0 <= nx < N and 0 <= ny <M and arr[ny][nx] < arr[y][x] and visited[y][x] == 0:
            print(f'{sad_cnt}번째, ny : {ny}, nx: {nx} 가 다음에 바뀔 예정이에요!{"-"* 20}')
            sad_cnt +=1
            visited[y][x] = 1
            for i in visited:
                print(i)
            print('-' * 50)
            dfs(ny, nx)
            print(f'히히, 쓸모없어진 y : {y}와 x {x}를 다시 원래대로 바꿔버릴거에용')
            visited[y][x] = 0

dfs(0,0)
print(cnt)
print(sad_cnt)
```

```python
dp = [[-1] * N for _ in range(M)]

dy = (-1, 1, 0, 0)
dx = (0, 0, -1, 1)

sad_cnt = 0
def solve(y, x):
    global sad_cnt
    print(y, x)
    # 종료 조건
    if y == M-1 and x == N-1:
        print('앗! 종료에요! 너넨 전부 1이 되버려랏!!')
        return 1

    # 메모이제이션
    if dp[y][x] != -1:
        print(f'이런.. 이미 갔던데에요! dp[y][x] = {dp[y][x]}로 리턴할게요')
        return dp[y][x]

    dp[y][x] = 0
    for i in range(4):
        ny = y + dy[i]
        nx = x + dx[i]

        if not (0 <= ny < M and 0 <= nx < N):
            continue

        if arr[ny][nx] < arr[y][x]:
            sad_cnt +=1
            print(f'{sad_cnt}번째, ny : {ny}, nx: {nx} 가 다음에 바뀔 예정이에요!{"-"* 20}')
            for i in dp:
                print(i)
            print('-'*50)
            dp[y][x] += solve(ny, nx)
            print(f'리턴이 들어왔어요! y = {y}, x = {x}의 dp[y][x]는 {dp[y][x]} 에요!')
    print()
    return dp[y][x]



print(solve(0, 0))
print(sad_cnt)
```
