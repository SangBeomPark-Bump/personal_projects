```python
verbose = True
verboses = [True for _ in range(10)]
```

```python

raw_arr = """103000509
002109400
000704000
300502006
060000050
700803004
000401000
009205800
804000107""".split('\n')


arr = [list(map(int, list(row))) for row in raw_arr]
```

```python
def sudoku_determine(y, x, number):
    global arr
    if arr[y][x] != 0:
        return False

    for i in range(9):
        if arr[y][i] == number:
            return False
        if arr[i][x] == number:
            return False
        
    for j in range(3):
        for i in range(3):
            if arr[j + y//3][i + x//3] == number:
                return False
    
    return True
```

```python
recursion = 0
def dfs(y, x, number):
    # if verboses[0]:
    #     print(f'x {x}, y {y}, number {number}')
    global recursion

    if number == 10:
        dfs(y, x+1, 1)
        return

    if x == 9:
        dfs(y+1, 0, 1)
        return

    if y == 9:
        return
    
    recursion +=1
    if recursion > 10000:
        return
    if sudoku_determine(y, x, number):
        arr[y][x] = number
        if verbose:
            print(f'y = {y}, x = {x}, number = {number}')
            for i in arr:
                print(i)
        if x == 8:
            dfs(y+1, 0, 1)
        else:
            dfs(y, x+1, 1)
        arr[y][x] = 0
    dfs(y,x,number+1)
```
