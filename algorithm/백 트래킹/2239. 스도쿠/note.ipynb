```python
def sudoku_determine(y, x, number):
    global arr
    if arr[y][x] != 0:
        return False

    for i in range(9):
        if arr[y][i] == number:
            return False
        if arr[i][x] == number:
            return False
        
    for j in range(3):
        for i in range(3):
            if arr[j + y//3][i + x//3] == number:
                return False
    
    return True
```

```python
recursion = 0
def dfs(x, y, number):
    # if verboses[0]:
    #     print(f'x {x}, y {y}, number {number}')
    global recursion

    if x == 9:
        return

    if y == 9:
        return

    if number == 10:
        return
    

    for i in range(1, 10):
        if sudoku_determine(y, x, i):
            arr[y][x] = i
            if verbose:
                print(f'y = {y}, x = {x}, number = {i}')
                for i in arr:
                    print(i)
            recursion +=1
            if recursion >1000:
                return
            
            dfs(x+1, y, 1)
            dfs(x, y+1, 1)
    
    dfs(x+1, y, 1)
    dfs(x, y+1, 1)


    # dfs(x, y, number+1)
    # number = 1
    # if arr[y][x] == 0:
    #     return
    # dfs(x+1, y, number)
    # x = 0
    # dfs(x, y+1, number)
    # y = 0

    arr[y][x] = 0
    # dfs(x, y+1)
```

```python

def sad_dfs(x, y, number):
    if verboses[0]:
        print(f'x {x}, y {y}, number {number}')
    global recursion

    if number == 10:
        return
    if x == 3 or y == 3:
        temp_min = 0
        for row in arr:
            temp_min = min(temp_min, min(row))

        if temp_min != 0:
            print(*arr)
        return
    

    if sudoku_determine(y, x, number):
        arr[y][x] = number
        if verbose:
            print(f'y = {y}, x = {x}, number = {number}')
            for i in arr:
                print(i)

    recursion +=1
    if recursion >1000:
        return

    sad_dfs(x, y, number+1)
    number = 1
    if arr[y][x] == 0:
        return
    sad_dfs(x+1, y, number)
    x = 0
    sad_dfs(x, y+1, number)
    y = 0

    arr[y][x] = 0
    # dfs(x, y+1)
```
