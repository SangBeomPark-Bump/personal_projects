```python
import pandas as pd
```

```python
sad =  pd.read_csv(path + 'data/'+'train_words0.csv', index_col=0)
```

```python
responses =  sad.map(lambda x: x.split('### Response:\n')[-1])
inputs = sad.map(lambda x: x.split('### Response:\n')[0].split('### Input:\n')[-1])
```

```python
sad_sad =  pd.concat([inputs, responses], axis=1)
sad_sad.columns = ['input', 'response']
```

```python
n = 3
num= n * 10
so_sad_sad = sad_sad.iloc[num : num+10]

for idx, row in so_sad_sad.iterrows():
    print(f'{idx + 1}번째 질문과 답변')
    print(row['input'],  end='')
    print(row['response'])
    print()
    print('-' * 80)
    print()
```

```python
sad_responses_mask = responses.iloc[:, 0].map( lambda x: len(x) > 70 )
```

```python
sad_sad_sad = sad_sad[sad_responses_mask]
```
