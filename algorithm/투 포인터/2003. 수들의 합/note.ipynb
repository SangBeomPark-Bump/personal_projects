```python
verbose = True
```

```python
N, M = map(int, "10 5".split())
arr =  list(map(int, "1 2 3 4 2 5 3 1 1 2".split())) + [0]
```

```python
start = 0
end = 0

sum = arr[end]
result = 0

iter = 0
while start != N and end != N:
    iter +=1
    if sum < M:
        end +=1
        sum += arr[end]
    elif sum > M :
        sum -= arr[start]
        start +=1
    else:
        result +=1
        end +=1
        sum +=arr[end]
    if verbose:
        print(f'start: {start}, end : {end}, sum : {sum}')
        print('-' * 100)
if verbose:
    print(f'반복횟수 : {iter}')
print(result)
```
